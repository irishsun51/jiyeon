```python
# [퀘스트 1번: 은행계좌 만들기]
# Q1. Account 클래스 : 은행에 가서 계좌를 개설하면 은행이름, 예금주, 계좌번호, 잔액이 설정됩니다. Account 클래스를 생성한 후 생성자(hint: 매직메서드...!!)를 구현해보세요. 생성자에서는 예금주와 초기 잔액만 입력 받습니다. 은행이름은 SC은행으로 계좌번호는 3자리-2자리-6자리 형태로 랜덤하게 생성됩니다. (은행이름: SC은행, 계좌번호: 111-11-111111)
# def __init__(self, name, balance): 형태를 사용해.
# 계좌번호 랜덤 생성을 위해 random 모듈의 randint를 활용해.

# Q2. 클래스 변수 : 클래스 변수를 사용해서 Account 클래스로부터 생성된 계좌 객체의 개수를 저장하세요.
# 클래스 바로 아래에 account_count = 0처럼 선언
# 생성자(__init__) 내부에서 Account.account_count += 1을 해줘야 객체가 생길 때마다 숫자가 올라감.

# Q3. 클래스 변수 출력 : Account 클래스로부터 생성된 계좌의 개수를 출력하는 get_account_num() 메서드를 추가하세요.
# 객체가 없어도 호출할 수 있도록 @classmethod 데코레이터를 사용

# Q4. 입금 메서드 : Account 클래스에 입금을 위한 deposit 메서드를 추가하세요. 입금은 최소 1원 이상만 가능합니다.
# if amount >= 1: 조건문을 사용하여 잘못된 값이 들어오지 않게

# Q5. 출금 메서드 : Account 클래스에 출금을 위한 withdraw 메서드를 추가하세요. 출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.
# if self.balance >= amount: 조건을 통해 잔액 부족 상황을 체크

# Q6. 정보 출력 메서드 : Account 인스턴스에 저장된 정보를 출력하는 display_info() 메서드를 추가하세요. 잔고는 세자리마다 쉼표를 출력하세요.
# (은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)
# 숫자 세 자리 콤마는 파이썬의 포맷팅 문법인 "{:,}".format(value) 또는 f"{value:,}"를 사용

# Q7. 이자 지급하기 : 입금 횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가되도록 코드를 변경해보세요.
# 입금 횟수를 기록할 인스턴스 변수(self.deposit_count)가 필요합니다.
# if self.deposit_count % 5 == 0:을 사용하여 5배수일 때만 이자를 계산

# Q8. 여러 객체 생성 : Account 클래스로부터 3개 이상 인스턴스를 생성하고 생성된 인스턴스를 리스트에 저장해보세요.
# accounts = [] 리스트를 만들고 accounts.append(Account(...))를 세 번 반복

# Q9. 객체 순회 반복문을 통해 리스트에 있는 객체를 순회하면서 잔고가 100만원 이상인 고객의 정보만 출력하세요.
# for acc in accounts:로 순회하며 if acc.balance >= 1000000:인 경우에만 acc.display_info()를 호출

# Q10. 입금과 출금 내역이 기록되도록 코드를 업데이트 하세요.
# (입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드를 추가하세요.)
# 리스트(self.deposit_log = [], self.withdraw_log = [])를 만들어 append로 내역을 쌓아두기
```

```python
import random

class Account:
    # Q2. 클래스 변수: 생성된 계좌의 개수를 저장
    account_count = 0

    def __init__(self, name, balance):
        # Q1. 생성자: 예금주와 초기 잔액 설정
        self.name = name
        self.balance = balance
        self.bank = "SC은행"

        # 계좌번호 랜덤 생성 (3자리-2자리-6자리)
        num1 = random.randint(0, 999)
        num2 = random.randint(0, 99)
        num3 = random.randint(0, 999999)

        # zfill을 사용하여 자리수를 맞춤 (예: 1 -> 001)
        self.account_number = f"{str(num1).zfill(3)}-{str(num2).zfill(2)}-{str(num3).zfill(6)}"

        # Q7을 위한 입금 횟수 기록
        self.deposit_count = 0

        # Q10을 위한 입출금 내역 리스트
        self.deposit_log = []
        self.withdraw_log = []

        # 객체가 생성될 때마다 클래스 변수 증가
        Account.account_count += 1

    # Q3. 클래스 변수 출력 메서드
    @classmethod
    def get_account_num(cls):
        print(f"현재 생성된 계좌 개수: {cls.account_count}개")

    # Q4 & Q7. 입금 메서드 (이자 지급 로직 포함)
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_log.append(amount) # Q10 내역 기록
            self.deposit_count += 1

            # Q7. 입금 횟수가 5회가 될 때마다 1% 이자 지급
            if self.deposit_count % 5 == 0:
                interest = self.balance * 0.01
                self.balance += int(interest)
        else:
            print("1원 이상만 입금 가능합니다.")

    # Q5. 출금 메서드
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            self.withdraw_log.append(amount) # Q10 내역 기록
        else:
            print("잔액이 부족합니다.")

    # Q6. 정보 출력 메서드 (세자리 쉼표 포함)
    def display_info(self):
        print(f"은행이름: {self.bank}")
        print(f"예금주: {self.name}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔고: {self.balance:,}원") # :, 를 사용하여 천 단위 콤마 표시

    # Q10. 입출금 내역 출력 메서드
    def deposit_history(self):
        print(f"--- {self.name}님의 입금 내역 ---")
        for log in self.deposit_log:
            print(f"입금: {log:,}원")

    def withdraw_history(self):
        print(f"--- {self.name}님의 출금 내역 ---")
        for log in self.withdraw_log:
            print(f"출금: {log:,}원")

# ---------------------------------------------------------
# Q8. 여러 객체 생성 및 리스트 저장
accounts = []
acc1 = Account("파이썬", 10000)
acc2 = Account("자바", 2000000)
acc3 = Account("C언어", 500000)

accounts.append(acc1)
accounts.append(acc2)
accounts.append(acc3)

# Q9. 객체 순회 및 조건부 출력 (잔고 100만원 이상)
print("--- 잔고 100만원 이상 고객 정보 ---")
for acc in accounts:
    if acc.balance >= 1000000:
        acc.display_info()
        print("-" * 20)
```
